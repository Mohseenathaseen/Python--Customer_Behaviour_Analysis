```python
import numpy as np 
import pandas as pd 
```

```python
# reading the customer data 
df= pd.read_csv('customer_shopping_behavior (1).csv')
df.head(2)

```

```python
#df.describe-> give statistical summary of numeric data and if u want to categorial column then 
df.describe(include='all')
```

```python
# checking the missing values
df.isnull().sum()
```

```python
# how to r']=df.groupby('replace null values, median value is robust to outliers .so median for each category will be replaced with null values 
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
# checking any null value 
df.isnull().sum()
```

```python
#checking column name with snake casing i.e name should be in lower case with single word through underscore 
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
# checking the columns 
df.columns
df= df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels) 
```

```python
# create new column for purchase frequency day-wise 
frequency_mapping ={
    'Fortnightly':14,
     'Weekly':7,
     'Monthly':30,
     'Quarterly':90,
     'Annually': 365,
    'Bi-Weekly':14,
    'Every 3 months':90 
}     
```

```python
df['frequency_of_purchases_days']= df['frequency_of_purchases'].map(frequency_mapping)
```
